```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/cleaned_innings_results_data.csv")
```

```python
# Convert 'Start Date' to datetime format
df['Start Date'] = pd.to_datetime(df['Start Date'])
```

Data analysis on Ahmedabad stadium

```python
# Filter rows where the venue is "Ahmedabad"
ahmedabad_rows = df[df['Ground'] == 'Ahmedabad']
```

```python
ahmedabad_rows.head()
```

```python
# Convert 'Start Date' to datetime format
df['Start Date'] = pd.to_datetime(df['Start Date'])

# Select only innings 1
innings1_df = df[df['Inns'] == 1].copy()

# Rename the DataFrame (you can replace 'innings1_df' with your preferred name)
innings1_df = innings1_df.reset_index(drop=True)  # Resetting index for a clean index

# # Display the resulting DataFrame
# print(innings1_df)
```

```python
innings1_df.head()
```

```python
# Create a copy of innings1_df with matches played in Ahmedabad
ahmedabad_innings1_df = innings1_df[innings1_df['Ground'] == 'Ahmedabad'].copy()

# # Filter matches with 'Start Date' after the year 2020
# ahmedabad_innings1_df = ahmedabad_innings1_df[ahmedabad_innings1_df['Start Date'].dt.year > 2000]

# Plotting
plt.figure(figsize=(10, 6))
ahmedabad_innings1_df['Team'].value_counts().plot(kind='bar', color='blue')
plt.title('Number of Matches Played by Each Team in Ahmedabad (Inning 1)')
plt.xlabel('Team')
plt.ylabel('Number of Matches')
plt.show()
```

```python
ahmedabad_innings1_df.head()
```

```python
# Sort by 'Start Date' in descending order
ahmedabad_innings1_df = ahmedabad_innings1_df.sort_values(by='Start Date', ascending=False)
ahmedabad_innings1_df.head(30)
```

```python
import seaborn as sns

# Create a copy of innings1_df with matches played in Ahmedabad
ahmedabad_innings1_df = innings1_df[innings1_df['Ground'] == 'Ahmedabad'].copy()

# Plotting
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Team', y='Score', hue='Opposition', data=ahmedabad_innings1_df, s=100, palette='viridis')
plt.title('Team vs Opposition vs Score in Ahmedabad (Inning 1)')
plt.xlabel('Team')
plt.ylabel('Score')
plt.show()
```
